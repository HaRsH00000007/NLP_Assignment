```python
import pandas as pd
import spacy
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from spacy.lang.en.stop_words import STOP_WORDS

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")
```

# **1. Preprocessing:**

# **1.1 Text Cleaning (Lowercasing, Removing Punctuation, Stopwords)** 

```python
def clean_text(text):
    if pd.isna(text):  # Handle missing values
        return ""

    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z0-9\s]', '', text)  # Remove special characters & punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    cleaned_text = " ".join([word for word in words if word not in stop_words])
    
    return cleaned_text
```

```python
# Apply cleaning to relevant columns
df['facts'] = df['facts'].apply(clean_text)
df['first_party'] = df['first_party'].apply(clean_text)
df['second_party'] = df['second_party'].apply(clean_text)
df['decision_type'] = df['decision_type'].apply(clean_text)
df['disposition'] = df['disposition'].apply(clean_text)
df['issue_area'] = df['issue_area'].apply(clean_text)
```

# **1.2 Tokenization & POS Tagging**

```python
def tokenize_and_tag(text):
    if pd.isna(text) or text.strip() == "":
        return []
    
    doc = nlp(text)  
    tokens_pos = [(token.text, token.pos_) for token in doc]  # Token-POS pairs
    return tokens_pos

# Apply POS tagging on the 'facts' column (main text for task extraction)
df['facts_pos'] = df['facts'].apply(tokenize_and_tag)
```

```python
def sentence_split_parse(text):
    if pd.isna(text) or text.strip() == "":
        return []

    doc = nlp(text)
    
    # Extract sentences
    sentences = [sent.text for sent in doc.sents]

    # Extract dependency relations
    dependency_parsing = []
    for token in doc:
        dependency_parsing.append((token.text, token.dep_, token.head.text))

    return sentences, dependency_parsing
```

# **1.3 Sentence Splitting & Dependency Parsing**

```python
# Apply function to extract sentences and dependencies
df[['facts_sentences', 'facts_dependencies']] = df['facts'].apply(lambda x: pd.Series(sentence_split_parse(x)))
```

# **2. Implement Task Extraction Logic**

# **Step 2.1: Extracting Verbs, Assignment Phrases & Time Indicators**

```python
import spacy
from collections import Counter

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Initialize lists to collect extracted words
all_verbs = []
all_assignment_phrases = []
all_time_indicators = []

# Common modal verbs & keywords indicating assignment
common_assignment_keywords = {"has to", "needs to", "should", "must", "is required to"}

# Common time indicators
common_time_keywords = {"by", "before", "within", "until", "at", "on", "tomorrow", "today"}

def extract_keywords(sentences):
    for sentence in sentences:
        doc = nlp(sentence)
        
        # Extract verbs
        verbs = [token.lemma_.lower() for token in doc if token.pos_ == "VERB"]
        all_verbs.extend(verbs)
        
        # Extract assignment-related phrases
        for token in doc:
            if token.text.lower() in common_assignment_keywords:
                all_assignment_phrases.append(token.text.lower())
        
        # Extract time-related words
        for token in doc:
            if token.text.lower() in common_time_keywords:
                all_time_indicators.append(token.text.lower())

# Apply function to extract keywords from 'facts_sentences'
df["facts_sentences"].dropna().apply(lambda x: extract_keywords(x) if isinstance(x, list) else None)

# Get the most common words for each category
verbs_counter = Counter(all_verbs).most_common(30)  # Get top 30 verbs
assignment_counter = Counter(all_assignment_phrases).most_common(15)
time_counter = Counter(all_time_indicators).most_common(15)

# Convert to sets for fast lookup
task_verbs = {word for word, _ in verbs_counter}
assignment_phrases = {word for word, _ in assignment_counter}
time_indicators = {word for word, _ in time_counter}

print("Extracted Verbs:", task_verbs)
print("Extracted Assignment Phrases:", assignment_phrases)
print("Extracted Time Indicators:", time_indicators)
```

```python
def extract_tasks(sentences):
    identified_tasks = []
    
    for sentence in sentences:
        doc = nlp(sentence)
        words = {token.text.lower() for token in doc}  # Lowercased words in sentence
        verbs = {token.lemma_.lower() for token in doc if token.pos_ == "VERB"}  # Extract verbs
        
        # Heuristic 1: Sentence starts with a verb (imperative form)
        if len(doc) > 1 and doc[0].pos_ == "VERB":
            identified_tasks.append(sentence)
            continue  
        
        # Heuristic 2: Contains extracted action verbs
        if verbs.intersection(task_verbs):
            identified_tasks.append(sentence)
            continue
        
        # Heuristic 3: Contains extracted assignment phrases
        if any(phrase in sentence.lower() for phrase in assignment_phrases):
            identified_tasks.append(sentence)
            continue
        
        # Heuristic 4: Contains extracted time constraints
        if words.intersection(time_indicators):
            identified_tasks.append(sentence)
            continue
        
    return identified_tasks

# Apply function to extract tasks
df["identified_tasks"] = df["facts_sentences"].apply(lambda x: extract_tasks(x) if isinstance(x, list) else [])
```

# **Step 2.2: Extract Task Details**|

Step 1: Extracting Responsibility (Who is Responsible?)

Using spaCy's Named Entity Recognition (NER) to detect PERSON entities (e.g., "Rahul", "John").

If no PERSON entity is found, then looking for pronouns (e.g., "he", "she", "we").


Step 2: Extracting Action (Verb + Object)

Using dependency parsing to find main verb and its direct object.

Example:

Sentence: "Rahul must file the appeal today."

Verb: "file", Object: "the appeal"

Extracted Action: "file the appeal"


Step 3: Extracting Deadline (Time Expression)

Using spaCy's DATE & TIME NER model to detect deadlines like "by 5 PM", "today", "within 24 hours".

Use regex for extra matching.

```python
import spacy
import re

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Function to extract responsibility, action, and deadline
def extract_task_details(sentences):
    task_details = []
    
    for sentence in sentences:
        doc = nlp(sentence)
        
        # Extract 'Who is Responsible?'
        responsible = None
        for ent in doc.ents:
            if ent.label_ == "PERSON":  # Check if entity is a person
                responsible = ent.text
                break
        
        # If no named person found, look for pronouns
        if not responsible:
            for token in doc:
                if token.text.lower() in {"he", "she", "we", "they", "you"}:
                    responsible = token.text
                    break

        # Extract 'What Action Needs to Be Done?'
        action = None
        for token in doc:
            if token.lemma_ in task_verbs:  
                # Get the direct object (what is being acted upon)
                obj = " ".join([child.text for child in token.children if child.dep_ == "dobj"])
                action = f"{token.text} {obj}" if obj else token.text
                break
        
        # Extract 'Deadline'
        deadline = None
        for ent in doc.ents:
            if ent.label_ in {"DATE", "TIME"}:
                deadline = ent.text
                break
        
        # Regex-based backup for time extraction
        time_regex = r"\b(by|before|within|until)\s+\d{1,2}(:\d{2})?\s*(AM|PM)?|today|tomorrow"
        match = re.search(time_regex, sentence, re.IGNORECASE)
        if match:
            deadline = match.group()
        
        # Store results
        task_details.append({
            "sentence": sentence,
            "responsible": responsible if responsible else "Unknown",
            "action": action if action else "Unknown",
            "deadline": deadline if deadline else "None"
        })
    
    return task_details

# Apply function to the identified tasks column
df["task_details"] = df["identified_tasks"].apply(lambda x: extract_task_details(x) if isinstance(x, list) else [])

# Convert to structured DataFrame
task_df = df.explode("task_details")
task_df = task_df.dropna(subset=["task_details"])  # Remove empty rows

# Extract details into separate columns
task_df["responsible"] = task_df["task_details"].apply(lambda x: x["responsible"])
task_df["action"] = task_df["task_details"].apply(lambda x: x["action"])
task_df["deadline"] = task_df["task_details"].apply(lambda x: x["deadline"])

# Drop raw task_details dictionary column
task_df.drop(columns=["task_details"], inplace=True)
```

# **3. Categorization:** 

# **Step 3.1: Implement Rule-Based Categorization**

```python
# Define categories based on extracted action verbs
task_categories = {
    "Legal Tasks": {"file", "appeal", "argue", "rule", "convict", "dismiss", "grant", "sue"},
    "Work Tasks": {"submit", "report", "complete", "review", "approve", "schedule"},
    "Academic Tasks": {"assignment", "exam", "study", "homework", "research"},
    "Household Chores": {"clean", "wash", "organize", "cook"},
    "Personal Tasks": {"buy", "call", "visit", "meet", "travel"}
}

# Function to assign category
def categorize_task(action):
    if not action or action == "Unknown":
        return "Uncategorized"

    for category, keywords in task_categories.items():
        for keyword in keywords:
            if keyword in action:
                return category
    return "Uncategorized"

# Apply categorization
task_df["category"] = task_df["action"].apply(categorize_task)
print(task_df[["action", "category"]].head(5))
```

# **Step 3.2: ML-Based Categorization Using Clustering**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Convert task actions into TF-IDF vectors
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(task_df["action"].fillna(""))

# Apply K-Means Clustering
num_clusters = 5  # Adjust based on dataset
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
task_df["cluster"] = kmeans.fit_predict(X)

# Print cluster assignments
print(task_df[["action", "cluster"]].head())
```

# **Step 1: Automatic Cluster Naming with LDA**

```python
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np

# Apply LDA on TF-IDF vectors
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda_topics = lda.fit_transform(X)

# Assign dominant topic per task
task_df["topic"] = np.argmax(lda_topics, axis=1)

# Get top words for each topic to name clusters
feature_names = vectorizer.get_feature_names_out()
topic_keywords = {}

for topic_idx, topic in enumerate(lda.components_):
    top_words = [feature_names[i] for i in topic.argsort()[:-6:-1]]  # Top 5 words
    topic_keywords[topic_idx] = ", ".join(top_words)

# Map topic to category names based on keywords
task_df["topic_name"] = task_df["topic"].map(topic_keywords)

print(task_df[["action", "topic_name"]].head())
```
